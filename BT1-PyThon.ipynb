```python
 import random

class MovableObject :
    def __init__(self,x,y,width,height):
        self.location = Pyvector(x,y)
        self.velocity = Pyvector(0.1,0.0)
        self.acceleration = Pyvector(0.002,0.00001)
        self.width = width
        self.height= height
        self.maxspeed = 2
    def update(self):
        self.velocity.add(self.acceleration);
        self.velocity.limit(self.maxspeed)
        self.location.add(self.velocity)
        self.checkEdges()
    
    def checkEdges(self):
        if self.location.x >= 100:
            i=self.location.y +5
            self.location=Pyvector(0,i)
        if self.location.y >= 100:
            
            self.location=Pyvector(0,0)
            
        
        
```
